```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
from tqdm import tqdm
import os
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

# CONFIGURATION
TRAIN_DIR = "/home/student/Downloads/copycat_fer_project/npdd_sl"
TEST_DIR  = "/home/student/Downloads/copycat_fer_project/test"
SAVE_DIR  = "/home/student/Downloads/copycat_fer_project/models"

EPOCHS = 25
BATCH_SIZE = 32
LR = 1e-4
NUM_CLASSES = 6
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

os.makedirs(SAVE_DIR, exist_ok=True)

# TRANSFORMS
transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=3),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# DATASETS
train_ds = datasets.ImageFolder(TRAIN_DIR, transform=transform)
test_ds  = datasets.ImageFolder(TEST_DIR, transform=transform)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False)

print("Class mapping:", train_ds.class_to_idx)

# MODEL (COPYCAT CNN)
model = models.vgg16(pretrained=True)
model.classifier[6] = nn.Linear(4096, NUM_CLASSES)
model = model.to(DEVICE)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LR)

# METRIC STORAGE
train_acc, test_acc = [], []
train_loss, test_loss = [], []

# TRAIN + TEST LOOP
for epoch in range(EPOCHS):
    # ---------------- TRAIN ----------------
    model.train()
    correct, total, running_loss = 0, 0, 0

    for x, y in tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS} [TRAIN]"):
        x, y = x.to(DEVICE), y.to(DEVICE)

        optimizer.zero_grad()
        out = model(x)
        loss = criterion(out, y)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, pred = out.max(1)
        total += y.size(0)
        correct += pred.eq(y).sum().item()

    tr_loss = running_loss / len(train_loader)
    tr_acc = 100 * correct / total

    # TEST 
    model.eval()
    correct, total, running_loss = 0, 0, 0

    with torch.no_grad():
        for x, y in test_loader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            out = model(x)
            loss = criterion(out, y)

            running_loss += loss.item()
            _, pred = out.max(1)
            total += y.size(0)
            correct += pred.eq(y).sum().item()

    te_loss = running_loss / len(test_loader)
    te_acc = 100 * correct / total

    train_loss.append(tr_loss)
    test_loss.append(te_loss)
    train_acc.append(tr_acc)
    test_acc.append(te_acc)

    print(f"Epoch {epoch+1}: "
          f"Train Acc={tr_acc:.2f}% | Test Acc={te_acc:.2f}% | "
          f"Train Loss={tr_loss:.4f} | Test Loss={te_loss:.4f}")


# SAVE MODEL + METRICS
torch.save(model.state_dict(), f"{SAVE_DIR}/copycat_vgg16_npdd_sl.pth")

torch.save({
    "train_acc": train_acc,
    "test_acc": test_acc,
    "train_loss": train_loss,
    "test_loss": test_loss
}, f"{SAVE_DIR}/copycat_metrics.pth")

print("Copycat model & metrics saved.")

# PLOT ACCURACY GRAPH
epochs = range(1, EPOCHS + 1)

plt.figure(figsize=(8,5))
plt.plot(epochs, train_acc, label="Train Accuracy")
plt.plot(epochs, test_acc, label="Test Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.title("Copycat CNN – Train vs Test Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# PLOT LOSS GRAPH

plt.figure(figsize=(8,5))
plt.plot(epochs, train_loss, label="Train Loss")
plt.plot(epochs, test_loss, label="Test Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Copycat CNN – Train vs Test Loss")
plt.legend()
plt.grid(True)
plt.show()

# PCA CLUSTERING (TEST SET)

model.eval()
features, labels = [], []

with torch.no_grad():
    for x, y in test_loader:
        x = x.to(DEVICE)
        f = model.features(x)
        f = torch.flatten(f, 1)
        features.append(f.cpu().numpy())
        labels.append(y.numpy())

features = np.concatenate(features)
labels = np.concatenate(labels)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(features)

plt.figure(figsize=(8,6))
for cls in np.unique(labels):
    idx = labels == cls
    plt.scatter(X_pca[idx,0], X_pca[idx,1], label=str(cls), alpha=0.6)

plt.legend(title="Class")
plt.title("Copycat CNN – PCA Clustering (Test Features)")
plt.grid(True)
plt.show()
```
